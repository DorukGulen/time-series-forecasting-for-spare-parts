```python
import pandas as pd


df["Yaratma tarihi"] = pd.to_datetime(df["Yaratma tarihi"], errors="coerce")


ilk_tarih = df.groupby("GuncelMalzeme")["Yaratma tarihi"].min()
eligible_materials = ilk_tarih[ilk_tarih >= pd.Timestamp("2021-01-01")].index


mask = (df["Yaratma tarihi"] > "2024-06-30") & (df["Yaratma tarihi"] <= "2025-06-30")
df_win = df.loc[mask & df["GuncelMalzeme"].isin(eligible_materials)].copy()


frekans = (
    df_win.groupby("GuncelMalzeme")["Yaratma tarihi"]
    .apply(lambda x: x.dt.to_period("M").nunique())
    .rename("Frekans")
)


frekans = frekans.reindex(eligible_materials, fill_value=0).reset_index()

# --- 4) Kategori fonksiyonu ---
def kategori(x: int) -> str:
    if x == 0:
        return "0"
    elif 1 <= x <= 2:
        return "1-2"
    elif 3 <= x <= 5:
        return "3-5"
    elif 6 <= x <= 8:
        return "6-8"
    elif 9 <= x <= 11:
        return "9-11"
    elif x == 12:
        return "12"
    else:
        return "Diğer"


frekans["Frekans_Kategori"] = frekans["Frekans"].apply(kategori)

# frekans dataframe'i: ["GuncelMalzeme", "Frekans", "Frekans_Kategori"]

```

```python
import pandas as pd
import numpy as np

# 1) Liste: GuncelMalzeme
a = b["GuncelMalzeme"].astype(str).str.strip().unique()

# 2) Kaynak df'yi hazırlama
df_src = df.copy()
key_col = "malzeme_kodu" if "malzeme_kodu" in df_src.columns else "GuncelMalzeme"
df_src[key_col] = df_src[key_col].astype(str).str.strip()

# 3) Yıl / Ay çıkar
if "Yaratma tarihi" in df_src.columns:
    df_src["Yaratma tarihi"] = pd.to_datetime(df_src["Yaratma tarihi"], errors="coerce")
    df_src["yil"] = df_src["Yaratma tarihi"].dt.year
    df_src["ay"]  = df_src["Yaratma tarihi"].dt.month
elif not {"yil","ay"}.issubset(df_src.columns):
    raise KeyError("Aylık gruplayabilmek için 'Yaratma tarihi' ya da 'yil/ay' kolonları gerekli.")

# 4) Sadece listede olan malzemeleri al
df_filt = df_src[df_src[key_col].isin(a)].copy()

# 5) Aylık toplam
monthly_sum = (
    df_filt.groupby([key_col, "yil", "ay"], as_index=False)["Sipariş miktarı"]
    .sum()
    .rename(columns={"Sipariş miktarı": "siparis_miktari_sum"})
)

# 6) Eksik ayları 0 ile doldur
# Her malzeme ve yıl için 1–12 ay arası tüm kombinasyonları oluştur
full_index = (
    pd.MultiIndex.from_product(
        [monthly_sum[key_col].unique(), monthly_sum["yil"].unique(), range(1, 13)],
        names=[key_col, "yil", "ay"]
    )
)

# Reindex ile eksik kombinasyonları 0 yap
monthly_sum = (
    monthly_sum
    .set_index([key_col, "yil", "ay"])
    .reindex(full_index, fill_value=0)
    .reset_index()
)

# 7) Sonuç sıralama
monthly_sum = monthly_sum.sort_values([key_col, "yil", "ay"]).reset_index(drop=True)

monthly_sum.head(20)
```
