```python
monthly = (
    df.groupby(["GuncelMalzeme", "YearMonth"], as_index=False)["Sipariş miktarı"]
      .sum()
      .rename(columns={"Sipariş miktarı": "siparis_miktari"})
)

all_months = pd.period_range(df["Yaratma tarihi"].min(), df["Yaratma tarihi"].max(), freq="M")
all_materials = df["GuncelMalzeme"].unique()
full_idx = pd.MultiIndex.from_product([all_materials, all_months], names=["GuncelMalzeme", "YearMonth"])

monthly_full = (
    monthly.set_index(["GuncelMalzeme", "YearMonth"])
           .reindex(full_idx, fill_value=0)
           .reset_index()
)
```

```python
monthly_full=monthly_full.rename(columns={"GuncelMalzeme": "malzeme_kodu"})
```

```python
siparis_sayilari = monthly_full[monthly_full["siparis_miktari"] > 0]["malzeme_kodu"].value_counts()

az_gecen_malzeme_kodlari = siparis_sayilari[siparis_sayilari < 5].index.tolist()

df_cikartilan = pd.DataFrame({"malzeme_kodu": az_gecen_malzeme_kodlari})
df_cikartilan.to_excel("5den_az_siparis_gelen_malzemeler.xlsx", index=False)

monthly_full = monthly_full[~monthly_full["malzeme_kodu"].isin(az_gecen_malzeme_kodlari)]
```

```python
# Tarihi dönüştür
dolar["ds"] = pd.to_datetime(dolar["Tarih"], dayfirst=True, errors="coerce")
dolar["YearMonth"] = dolar["ds"].dt.to_period("M")

# "Şimdi" kolonunu numeriğe çevir
dolar["Şimdi"] = (dolar["Şimdi"].astype(str)
                  .str.replace(".", "", regex=False)   # binlik ayırıcıları sil
                  .str.replace(",", ".", regex=False)) # ondalık için nokta yap
dolar["Şimdi"] = pd.to_numeric(dolar["Şimdi"], errors="coerce")

# Aylık ortalama dolar
dolar_monthly = (
    dolar.groupby("YearMonth", as_index=False)["Şimdi"].mean()
)

# Merge et
monthly_full = monthly_full.merge(dolar_monthly, on="YearMonth", how="left")
```

```python
# Tarihi datetime'a çevir
tüfe["ds"] = pd.to_datetime(tüfe["ds"], dayfirst=True, errors="coerce")
tüfe["YearMonth"] = tüfe["ds"].dt.to_period("M")

# TÜFE yıllık değerlerini sayıya çevir
tüfe["tüfe yıllık"] = (
    tüfe["tüfe yıllık"]
    .astype(str)                      # önce string
    .str.replace(",", ".", regex=False)  # , yerine . koy
    .str.replace(" ", "", regex=False)   # boşlukları temizle
)
tüfe["tüfe yıllık"] = pd.to_numeric(tüfe["tüfe yıllık"], errors="coerce")

# Aylık ortalama TÜFE hesapla
tufe_monthly = (
    tüfe.groupby("YearMonth", as_index=False)["tüfe yıllık"].mean()
         .rename(columns={"tüfe yıllık": "tufe"})
)

# Merge et
monthly_full = monthly_full.merge(tufe_monthly, on="YearMonth", how="left")

```

```python
import pandas as pd

# Tüm ayları üret (2021-01 → 2025-06)
all_months = pd.period_range("2021-01", "2025-06", freq="M")

# Kampanya DataFrame'inizi hazırlayın
kampanya = otob_kampanya.copy()
kampanya["Bşl.tarihi"] = pd.to_datetime(kampanya["Bşl.tarihi"], dayfirst=True, errors="coerce")
kampanya["YearMonth"] = kampanya["Bşl.tarihi"].dt.to_period("M")

# Tüm malzemeler
kampanya["GuncelMalzeme"] = kampanya["GuncelMalzeme"].astype(str)
all_materials = kampanya["GuncelMalzeme"].unique()

# Malzeme x Ay kombinasyonları
full_idx = pd.MultiIndex.from_product([all_materials, all_months], names=["GuncelMalzeme", "YearMonth"])
kampanya_full = pd.DataFrame(index=full_idx).reset_index()

# Kampanya olan (Malzeme, YearMonth) seti
kampanya_pairs = set(zip(kampanya["GuncelMalzeme"], kampanya["YearMonth"]))

# Kampanya flag (1/0) ekle
kampanya_full["kampanya"] = kampanya_full.apply(
    lambda row: 1 if (row["GuncelMalzeme"], row["YearMonth"]) in kampanya_pairs else 0,
    axis=1
)

# Excel çıktısı
kampanya_full.to_excel("kampanya_full.xlsx", index=False)
print("✅ Kampanya full tablo 2021-01 → 2025-06 arası oluşturuldu.")
```
